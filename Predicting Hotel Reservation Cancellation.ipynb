# MISSING VALUE HANDLING

```python
dict = {'children':0, 'agent':0, 'company':0}
df.fillna(dict, inplace=True)
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
data = df.loc[~((df.adults == 0) & (df.children == 0) & (df.babies == 0))]
print(data)
```

```python
data.reservation_status_date = pd.to_datetime(data.reservation_status_date)
```

# INPUT SPLIT

```python
X = data.drop(columns = ["arrival_date_year", "arrival_date_week_number", "is_canceled", 'company', "arrival_date_month", 'reserved_room_type',"customer_type", 'reservation_status_date'])
y = data["is_canceled"]
```

# TRAIN TEST SPLIT AND MODEL TRAINING

```python
X.info()
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify = y)
def train(model, X, y) :
    
    #model fitting
    model.fit(X_train, y_train)
    
    
    #predicting the result
    pred = model.predict(X_test)
    
    
    #cross validation
    cv_score = cross_val_score(model, X, y, cv=5)
    
    
    print("Results")
    print("CV Score:", np.mean(cv_score))
```

```python
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier(n_neighbors=5)
train(knn, X, y)
print("score:", knn.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
train(clf, X, y)
print("score:", clf.score(X_test, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
train(dtc, X, y)
print("score:", dtc.score(X_test, y_test))
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
train(logisticRegr, X, y)
print("score:", logisticRegr.score(X_test, y_test))
```
